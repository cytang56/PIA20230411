read and show an image (graylevel/color)

```python
import cv2
import matplotlib.pyplot as plt

#img = cv2.imread('Lenna.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.imread('Lenna.jpg')
plt.imshow(img)
```

改變 BGR 為 RGB

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

影像處理: 取區塊取區塊

```python
# Crop image
cropped_image = img[100:300, 100:300]

plt.imshow(cropped_image)
plt.show()

# Crop image
cropped_image = img[200:250]

plt.imshow(cropped_image)
plt.show()

# Crop image
cropped_image = img[:, 200:250]

plt.imshow(cropped_image)
plt.show()

# Crop image
cropped_image = img.copy()
cropped_image[100:250, 100:250,:] = 255 - cropped_image[100:250, 100:250,:]

plt.imshow(cropped_image)
plt.show()


# image: color
image = img[:, 200:250, :1]

plt.imshow(image)
plt.show()
```
